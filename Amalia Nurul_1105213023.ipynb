6. Buatlah program yang dapat mengecek input kata pertama merupakan anagram untuk input kata kedua! [10 POINTS]

```python
class hewan:
  def __init__(self, alur, ular):
    self.alur = alur
    self.ular = ular

  def info(self):
    return f"{self.alur} dan {self.ular} adalah anagram"
kalimat2 = hewan("ular", "alur")
print(kalimat2.info())
```

```python
class hewan:
  def __init__(self, alur, liar):
    self.alur = alur
    self.liar = liar
  def info(self):
    return f"{self.alur} dan {self.liar} bukan anagram"
kalimat2 = hewan("alur", "liar")
print(kalimat2.info())
```
